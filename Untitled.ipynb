```python
import numpy as np
import random
from numpy import random
Sudoku = random.randint(1, 10, size=(9, 9))

#Sudoku = [[5, 3, 0, 0, 7, 0, 0, 0, 0],
#          [6, 0, 0, 1, 9, 5, 0, 0, 0],
#          [0, 9, 8, 0, 0, 0, 0, 6, 0],
#          [8, 0, 0, 0, 6, 0, 0, 0, 3],
#          [4, 0, 0, 0, 8, 3, 0, 0, 1],
#          [7, 0, 0, 0, 2, 0, 0, 0, 6],
#          [0, 6, 0, 0, 0, 0, 2, 8, 0],
#          [0, 0, 0, 4, 1, 9, 0, 0, 5],
#          [0, 0, 0, 0, 8, 0, 0, 9, 7]]

def valid(su, number, i, j):
    for k in range(0, 9):
        if su[i][k] == number:
            return False
    for k in range(0, 9):
        if su[k][j] == number:
            return False
    i_start = (i // 3) * 3
    j_start = (j // 3) * 3
    for r in range(i_start, i_start + 3):
        for c in range(j_start, j_start + 3):
            if su[r][c] == number:
                return False
    return True

valid(Sudoku, 5, 0, 0)
```
